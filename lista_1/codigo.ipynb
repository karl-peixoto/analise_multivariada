```python
import pandas as pd
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.express as px
```

```python
dados = pd.read_csv('dados/questao_2_2.csv')
```

```python
fig = make_subplots(rows=4, cols=4)
cols = dados.columns[:-1]
for i in range(4):
    for j in range(4):
        if i != j:
            fig.add_trace(go.Scatter(y = dados[dados['sex'] == 'male'][cols[i]], x =dados[cols[j]] ,mode='markers', marker_color='blue'), row=i+1, col=j+1)
            fig.add_trace(go.Scatter(y = dados[dados['sex'] == 'female'][cols[i]], x =dados[cols[j]] ,mode='markers', marker_color='red'), row=i+1, col=j+1)
        else:
            fig.add_trace(go.Histogram(x = dados[cols[i]]), row=i+1, col=j+1)
        

    fig.update_xaxes(title_text= cols[i],row=4, col=i+1)
    fig.update_yaxes(title_text=cols[i],row=i+1, col=1)
fig.update_layout(autosize=False, width=800, height=600, showlegend=False)
fig.show()
```

```python
fig = px.scatter_3d(dados, x='x1', y='x2', z='x3', color='sex')
fig.write_html('graficos/questao_2_2_3d.html')
```
